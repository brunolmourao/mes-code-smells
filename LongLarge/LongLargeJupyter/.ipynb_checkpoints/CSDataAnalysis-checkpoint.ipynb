# Análise dos Dados do Processo de Identificação de Code Smells

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1) Large Class

### >> Sintomas: muitas linhas de código, muitos campos e muitos métodos.

### -- Projeto: commons-collections:

#### a) Visualização dos dados

```python
comLC = pd.read_csv('data/commonLC.csv')
comLC.head(10)
```

#### b) Descrição geral

```python
comLC.describe()
```

#### c) Histograma para o número de linhas

```python
plt.figure(figsize = (100, 10))

numClass = comLC.code_lines.value_counts().values
linesDist = comLC.code_lines.value_counts().index

sns.barplot(x = linesDist, y = numClass)
plt.xlabel('Numero de Linhas')
plt.ylabel('Numero de Classes')

plt.show()
```

#### d) Histograma para o número de campos

```python
plt.figure(figsize = (100, 10))

numClass = comLC.fields_num.value_counts().values
fieldsDist = comLC.fields_num.value_counts().index

g = sns.barplot(x = fieldsDist, y = numClass)
plt.xlabel('Numero de Campos')
plt.ylabel('Numero de Classes')

plt.show()
```

#### e) Histograma para o número de métodos

```python
plt.figure(figsize = (100, 10))

numClass = comLC.methods_num.value_counts().values
fieldsDist = comLC.methods_num.value_counts().index

g = sns.barplot(x = fieldsDist, y = numClass)
plt.xlabel('Numero de Métodos')
plt.ylabel('Numero de Classes')

plt.show()
```
