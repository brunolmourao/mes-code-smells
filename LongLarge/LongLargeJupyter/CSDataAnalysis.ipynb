### -- Avaliação commons vs xstream:

```python
def finalLC(dataset):
    smelly = []
    
    codeM = dataset.code_lines.median().round()
    fielM = dataset.fields_num.median().round()
    methM = dataset.methods_num.median().round()

    for cl in range(len(dataset)):
        actual = dataset.iloc[cl]
        if (actual.code_lines >= codeM and actual.fields_num >= fielM and actual.methods_num >= methM):
            smelly.append(actual.class_name)
    
    data = pd.DataFrame(smelly, columns=['Fedorentas'])
    per = (data.size / dataset.size) * 100
    print("Esse projeto eh {}% fedorento!!".format(per))
    return data
```

```python
finalLC(comLC)
```

```python
finalLC(xsLC)
```

### -- Avaliação commons vs. xstream:

```python
def finalLM(dataset):
    smelly = []
    
    codeM = dataset.code_lines.median().round()

    for cl in range(len(dataset)):
        actual = dataset.iloc[cl]
        if (actual.code_lines >= codeM):
            aux = [actual.class_name, actual.method_name]
            smelly.append(aux)
    
    data = pd.DataFrame(smelly, columns=['Classes', 'Fedorentos'])
    per = (data.size / dataset.size) * 100
    print("Esse projeto eh {}% fedorento!!".format(per))
    return data
```
