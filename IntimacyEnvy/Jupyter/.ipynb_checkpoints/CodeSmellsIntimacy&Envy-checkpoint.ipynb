#### c) Histograma para o número de usos externos

```python
plt.figure(figsize = (100, 10))

numClass = comII.num_external.value_counts().values
externalDist = comII.num_external.value_counts().index

sns.barplot(x = externalDist, y = numClass)
plt.xlabel('Numero de Usos Externos')
plt.ylabel('Numero de Classes')

for i in range(len(numClass)):
    plt.text(externalDist[i] - 1.05, numClass[i] + 0.8, str(numClass[i]), size = 20, fontweight='bold')

plt.show()
```
